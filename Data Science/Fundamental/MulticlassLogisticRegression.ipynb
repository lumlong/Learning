```python
from sklearn import datasets
import pandas as pd
import numpy as np
```

```python
# Load Iris dataset
X, y = datasets.load_iris(return_X_y=True, as_frame=True)
```

```python
def get_onehot(y):
    
    n_classes = len(np.unique(y))
    n_samples = len(y)
    
    y_onehot = np.zeros([n_samples, n_classes])
    for i, val in enumerate(y):
        y_onehot[i, val] = 1
        
    return y_onehot
```

```python
y_onehot = get_onehot(y)
y_onehot
```

```python
n_sample = X.shape[0]
n_feature = X.shape[1]

w = np.random.uniform(0, 0.1, [n_feature, 3])
b = 0.5


def net_input(X, w, b):
    net_input = X.dot(w.T) + b
    return net_input


def activation(z):
    # in y_predict_onehot = [3, 1.5,.5]
    # softmax(y_predict_onehot) = [.60, .30, .10]
    e_z = np.exp(z - np.max(z))
    return e_z / e_z.sum()


def cost(y_target, y_predict):
    cost = y_target * np.log(y_predict)
    return cost


def gradient(X, y, y_pred, learning_rate):
    
    error = y_pred - y
    
    # grad_w = error.T.dot(np.array(X)) / X.shape[0]
    grad_w = learning_rate * np.average((error * X.T).T, axis=0)
    # grad_b = error.sum() / X.shape[0]
    grad_b = learning_rate * np.average(error, axis=0)
    
    return grad_w, grad_b
```
